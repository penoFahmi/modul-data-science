```python
import pandas as pd
df = pd.read_csv("titanic.csv")
print ( df.head() )
```

```python
import matplotlib.pyplot as plt
# Line plot
plt . plot ([1 ,2 ,3 ,4] ,[10 ,20 ,25 ,30])
plt . title (" Line Plot ")
plt . show ()
# Bar chart
plt . bar (["A","B","C"] , [5 ,7 ,3])
plt . title ("Bar Chart ")
plt . show ()
# Histogram Age
df ["Age"]. hist ( bins =20 , color ="skyblue")
plt . title (" Distribusi Umur Penumpang ")
plt . show ()
# Scatter Age vs Fare
plt . scatter ( df ["Age"] , df ["Fare"] , alpha =0.5)
plt . title (" Scatter Age vs Fare ")
plt . xlabel ("Age")
plt . ylabel ("Fare")
plt . show ()
```

```python
import numpy as np
import seaborn as sns

# Countplot kategori Sex
sns.countplot(x="Sex", data=df)
plt.title("Distribusi Penumpang berdasarkan Jenis Kelamin")
plt.show()

# Boxplot Fare per Pclass
sns.boxplot(x="Pclass", y="Fare", data=df)
plt.title("Distribusi Harga Tiket per Kelas")
plt.show()

# Pilih hanya kolom numerik dari DataFrame
numeric_df = df.select_dtypes(include=np.number)

# Hitung korelasi pada DataFrame numerik
correlation_matrix = numeric_df.corr()

# Heatmap korelasi
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.title("Korelasi Antar Variabel Numerik")
plt.show()
```
